```python
import pandas as pd
import numpy as np
df = pd.read_csv('sample.csv')
df
df.shape
df.set_index('Example',inplace=True)
m=df.shape[0]
n=df.shape[1]
```

```python
def all(d):
    d=set(d)
    if len(d)==1:
#         print('yes')
        return True
def rem(d,s):
    c=0
    for i in range(len(d)):
        if(d[i]==s[i] or (d[i]=='?' and s[i]!='?') ):
                c+=1;
        else:
            return False
    return True
```
