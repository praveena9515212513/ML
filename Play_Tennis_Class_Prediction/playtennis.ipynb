```python
import pandas as pd
import numpy as np
df = pd.read_csv('playtennis.csv')
df
```

```python
data=df.loc[:9]
```

```python
data.set_index('Day',inplace=True)
```

```python
data
```

```python
fp=data[data['Class']=='P']
Pp=f.shape[0]/data.shape[0]
fn=data[data['Class']=='N']
Pn=fn.shape[0]/data.shape[0]
```

```python
col=data.columns
probs=[]
for i in range(data.shape[1]-1):
    d=data[col[i]]
    ds=set(d)
    p={}
    for j in ds:
        p[j]=[data[(data[col[i]]==j) & (data['Class']=='P')].shape[0]/fp.shape[0],
              data[(data[col[i]]==j) & (data['Class']=='N')].shape[0]/fn.shape[0]]
    probs.append(p)
probs
```

```python
tcs=df[10:]
tcs =np.array(tcs)
tcs = tcs[:,1:-1]
tcs
```

```python
for tc in tcs:
    ye=Pp
    no=Pn
    print(len(tc))
    for i in range(len(tc)):
        ye*=probs[i][tc[i]][0]
        no*=probs[i][tc[i]][1]

    print(ye,no)
    # probs[0]['overcast'][0]
    if(ye>no):
        print("He likes to play Tennis")
    else:
        print("He doesn't like to play Tennis")
```
